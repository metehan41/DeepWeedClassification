```python
import os
import shutil
import random
```

```python

# Define the paths
train_folder = '/content/drive/MyDrive/WeedClassification/data/weeds/train'
train_new_folder = '/content/drive/MyDrive/WeedClassification/data/weeds2/train2'
val_folder = '/content/drive/MyDrive/WeedClassification/data/weeds2/val'
test_folder = '/content/drive/MyDrive/WeedClassification/data/weeds2/test'
```

```python
# Create new folders if they don't exist
os.makedirs(train_new_folder, exist_ok=True)
os.makedirs(val_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

# List class folders in the train folder
class_folders = os.listdir(train_folder)
```

```python
# Define the split ratios
train_ratio = 0.7  # 70% for training
val_ratio = 0.15   # 15% for validation
test_ratio = 0.15  # 15% for testing
print(class_folders)
```

```python
for class_folder in class_folders[1:]:
    # Create new folders for each class in train, val, test
    os.makedirs(os.path.join(train_new_folder, class_folder), exist_ok=True)
    os.makedirs(os.path.join(val_folder, class_folder), exist_ok=True)
    os.makedirs(os.path.join(test_folder, class_folder), exist_ok=True)

    # List images/files in the current class folder
    files = os.listdir(os.path.join(train_folder, class_folder))

    # Shuffle the files to ensure randomness in splitting
    random.shuffle(files)

    # Calculate split indices
    total_files = len(files)
    train_split = int(total_files * train_ratio)
    val_split = int(total_files * (train_ratio + val_ratio))

    # Split the files into train, val, test sets
    train_files = files[:train_split]
    val_files = files[train_split:val_split]
    test_files = files[val_split:]

    # Move files to their respective folders
    for file in train_files:
        src = os.path.join(train_folder, class_folder, file)
        dst = os.path.join(train_new_folder, class_folder, file)
        shutil.copy(src, dst)

    for file in val_files:
        src = os.path.join(train_folder, class_folder, file)
        dst = os.path.join(val_folder, class_folder, file)
        shutil.copy(src, dst)

    for file in test_files:
        src = os.path.join(train_folder, class_folder, file)
        dst = os.path.join(test_folder, class_folder, file)
        shutil.copy(src, dst)
```
