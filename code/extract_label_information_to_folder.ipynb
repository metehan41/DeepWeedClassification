```python
import os
import shutil
import pandas as pd

# Path to your image folder and label CSV file
image_folder = '/content/drive/MyDrive/WeedClassification/data/images'
label_csv = '/content/drive/MyDrive/WeedClassification/data/labels/labels.csv'

# Create a directory to store the split images
output_folder = '/content/drive/MyDrive/WeedClassification/data/train'
os.makedirs(output_folder, exist_ok=True)

# Read the label CSV file
label_data = pd.read_csv(label_csv)

# Iterate through the label data
for index, row in label_data.iterrows():
    image_name = row['Filename']  # Replace 'Image_Name' with the column name in your CSV containing image names
    label = row['Label']  # Replace 'Label' with the column name in your CSV containing labels

    # Create subfolders based on labels if they don't exist
    label_folder = os.path.join(output_folder, str(label))
    os.makedirs(label_folder, exist_ok=True)

    # Path to the source image
    source_image_path = os.path.join(image_folder, image_name)

    # Path to the destination image
    destination_image_path = os.path.join(label_folder, image_name)

    # Move/copy the image to the respective label folder
    shutil.copy(source_image_path, destination_image_path)
```
